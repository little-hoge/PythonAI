```
from google.colab import files
from io import BytesIO
from matplotlib import pyplot as plt
import cv2

# アップロード
files = files.upload()
file_name = list(files.keys())[0]

# ファイル名表示テスト
print ('ファイル名', file_name)

# 読み込み
orig = cv2.imread(file_name)
im = cv2.cvtColor(orig, cv2.COLOR_BGR2RGB)

# 表示
plt.imshow(im)
plt.show()
```
